```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression


```

```python

# Create synthetic emails

spam_emails = [
    "Congratulations! You won a cash prize",
    "Claim your free lottery reward now",
    "You have been selected for a $5000 gift voucher",
    "Urgent!!! Update your bank details immediately",
    "Win an iPhone by clicking this link",
    "Limited offer, buy now and save 80%",
    "Your loan request is approved, click to verify",
    "Get cheap medicines without prescription",
    "You have an unclaimed reward waiting",
    "Exclusive deal just for you, hurry!"
] * 10

ham_emails = [
    "Please find the meeting agenda attached",
    "Let's schedule a call for tomorrow",
    "Your order has been shipped successfully",
    "Can you share the project files?",
    "Reminder: team meeting at 4 PM",
    "Thank you for your help",
    "Here is the report you requested",
    "Let's meet for lunch today",
    "The assignment deadline is extended",
    "Please review the attached document"
] * 10

emails = spam_emails + ham_emails
labels = [1]*100 + [0]*100  # 1 = spam, 0 = not spam

df = pd.DataFrame({"email": emails, "label": labels})

```

```python

# Data splitting

X_train, X_test, y_train, y_test = train_test_split(
    df["email"], df["label"],
    test_size=0.25,
    random_state=42
)


```

```python
# TF-IDF Vectorization

vectorizer = TfidfVectorizer(stop_words='english')
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)


# Logistic Regression (Sigmoid)

model = LogisticRegression()
model.fit(X_train_vec, y_train)

# Get sigmoid probabilities

probs = model.predict_proba(X_test_vec)[:, 1]   # probability of class 1 (spam)
preds = (probs >= 0.5).astype(int)


```

```python
# 6. Print in required output format
print("categories:['spam' 'not_spam']")

for i, (email, prob, pred) in enumerate(zip(X_test.tolist(), probs, preds), start=1):
    percent = prob * 100
    label = "Spam" if pred == 1 else "Not Spam"
    print(f"email{i}: Sigmoid output={prob:.4f} ({percent:.1f}%) -> Prediction: {label}")
```
