```python
from transformers import pipeline
```

```python
# Load summarization model (BERT-based via BART)
summarizer = pipeline(
    "summarization",
    model="facebook/bart-large-cnn"
)
```

```python
# Load Question Answering model (BERT)
qa_pipeline = pipeline(
    "question-answering",
    model="bert-large-uncased-whole-word-masking-finetuned-squad"
)
```

```python
# Load Question Answering model (BERT)
qa_pipeline = pipeline(
    "question-answering",
    model="bert-large-uncased-whole-word-masking-finetuned-squad"
)
```

```python
document = """
he Olympics are a global sporting event that bring together athletes from around the world.
They are held every four years and showcase a wide range of sports and competitions.
The Olympic Games promote unity, peace, and friendship among nations.
Athletes train for many years to represent their countries on this grand stage.
The Olympics include both individual and team events, testing strength, skill, and endurance.
Winning an Olympic medal is considered a great honor for any athlete.
The Olympic flag has five rings that symbolize the continents of the world.
The opening and closing ceremonies are colorful and culturally rich.
The Games inspire young people to take up sports and stay active.
Overall, the Olympics celebrate excellence, dedication, and the human spirit.
"""
```

```python
#Document summarization function
def summarize_text(text):
    summary = summarizer(
        text,
        max_length=80,
        min_length=30,
        do_sample=False
    )
    return summary[0]["summary_text"]
```

```python
print("ORIGINAL DOCUMENT:\n")
print(document)

print("\nGENERATED SUMMARY:\n")
summary = summarize_text(document)
print(summary)
```

```python
#Question answering function
def answer_question(context, question):
    result = qa_pipeline(
        question=question,
        context=context
    )
    return result["answer"]
```

```python
#ask question from document
question = input("Enter your question: ")
answer = answer_question(document, question)

print("\nANSWER:")
print(answer)
```
